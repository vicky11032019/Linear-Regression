```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('E:/HD/BA/ML/Projects Work/Linear model/LinearRegression-master/LinearRegressionTillCloud/Admission_Prediction.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['GRE Score'].fillna(df['GRE Score'].median(),inplace=True)

df
```

```python
df['TOEFL Score'].fillna(df['TOEFL Score'].median(),inplace=True)

df
```

```python
df['University Rating'].fillna(df['University Rating'].median(),inplace=True)

df
```

```python
X=df.iloc[ :,1:-1]

X
```

```python
y=df.iloc[ :,-1]
y
```

```python
X.columns
```

```python
for i in X.columns:
    plt.scatter(X[i],y)
    plt.xlabel(i)
    plt.ylabel('Chance of admission')
    plt.show()
    
```

```python
from sklearn.preprocessing import StandardScaler

std_feature=StandardScaler()

scaled_x=std_feature.fit_transform(X)

scaled_x

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=100)

```

```python
from sklearn import linear_model

lm = linear_model.LinearRegression()

lm.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score

score=r2_score(lm.predict(X_test),y_test)

score
```
